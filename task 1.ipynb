## BAR CHART


```python
import pandas as pd
import matplotlib.pyplot as plt
url = 'https://api.worldbank.org/v2/en/indicator/SP.POP.TOTL?downloadformat=csv'
df = pd.read_csv("worldpopulationdata.csv", skiprows=4)
df = df.melt(id_vars=["Country Name", "Country Code", "Indicator Name", "Indicator Code"], 
             var_name="Year", value_name="Population")
df.dropna(inplace=True)
df['Year'] = df['Year'].astype(int)
population_by_year = df.groupby('Year')['Population'].sum().reset_index()
plt.figure(figsize=(12, 6))
plt.bar(population_by_year['Year'].astype(str), population_by_year['Population'])
plt.xlabel('Year')
plt.ylabel('Total Population')
plt.title('Total World Population Over Years')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

## HISTOGRAM

```python
import pandas as pd
import matplotlib.pyplot as plt
url = 'http://api.worldbank.org/v2/en/indicator/SP.POP.TOTL?downloadformat=csv'
data = pd.read_csv("worldpopulationdata.csv", skiprows=4)
year = '1960'
population_data = data[['Country Name', year]].dropna()
plt.figure(figsize=(5,5))
plt.hist(population_data[year], bins=15, color='green',edgecolor='black')
plt.title(f'Population Distribution in {year}')
plt.xlabel('Population')
plt.ylabel('Number of Countries')
plt.show()
```
